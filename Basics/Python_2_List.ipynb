LIST - Most important python data type


What is List ?

List is a Data type where you can store multiple items under 1 name. More technically, lists acts like dynamic arrays which means you can add more items on the fly.

You can store multiple items of same data type or different to

Example :-

L = [ 20, 'Jessa' , 35.75 , [30,60,90] ]

Why Lists are required in programming ?

When you work with more data with similar data type at that time we use list to store them.

List is similar to array

Array Vs Lists



*   Fixed Vs Dynamic Size - We need to specify the array size while creating it where as lists are dynamic in nature lists are more flexible
*   Convenience - Arrays are homogenous like if we specify array as int then it should be integer array where as lists are hetrogenous that is you can add data of different data type


*   Speed of Execution - Python lists are slower in execution
*   Memory - Python lists occupy more space the aaray

```python
a = 2
print(id(a)) # prints the memory address of a
```

```python
L = [1,2,3]

print(id(L)) # memory address of list
print(id(L[0])) # memory addresses of the elements inside the list.
print(id(L[1]))
print(id(L[2]))
print(id(1)) # memory address where actual integer is stored
print(id(2))
print(id(3))
```

Creating a List

```python
# Empty
print([])

# 1D -> Homogenous
print([1,2,3,4,5])

# 2D -> nested -> hetrogenous
print([1,2,3,[4,5]])

# 3D -> Homogenous
print([[[1,2],[3,4]],[[5,6],[7,8]]])

# Hetrogenous
print([1,True,5.6,5+6j,'Hello'])

# Using Type conversion
print(list('hello'))
```

Accessing items from a List

```python
# Indexing
L = [1,2,3,4,5]
# positive -> Left to Right
print(L[0])
print(L[1])
print(L[2])

# Negative -> Right to Left
print(L[-1])
print(L[-2])

# 2D
L1 = [1,2,3,[4,5]]
print(L1[3][0]) # pos
print (L1[-1][-2]) # neg

# 3D
L2 = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L2[1][0][1]) # pos
print(L2[-1][-1][-2]) # neg
```

```python
# Slicing
L = [1,2,3,4,5,6]

print(L[0:3]) # positive
print(L[-3:-1]) # negative
print(L[-3:])
print(L[0::2])
print(L[-5:-2:2])
print(L[::-1])
```

Adding Items to a List

Editing items in a List

```python
L = [1,2,3,4,5]
# Editing with indexing
L[-1] = 500
# Editing with slicing
L[1:4] =[200,300,400]
print(L)
```

Deleting items in List

```python
# del
L = [1,2,3,4,5]
print(L)
# indexing
del L[-1]
print(L)
# slicing
del L[1:3]
print(L)
```

```python
# remove - ypu can remove by value
L = [1,2,3,4,5]
print(L)
L.remove(3)
print(L)
```

```python
# pop - by default it delets the last item
L = [1,2,3,4,5]
L.pop()
print(L)
L.pop(2)
print(L)
```

```python
# clear - delets all items
L = [1,2,3,4,5]
L.clear
print(L)
```

List Functions

```python
# len/min/max/sorted - universal
L = [2,1,5,7,6,3,4]
print(len(L))
print(min(L))
print(max(L))
print(sorted(L))
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,1,3,4,1,5]
L.count(1)
L.count(0)
```

```python
# index
L = [1,2,1,3,4,1,5]
L.index(2)
L.index(1) # prints index of first occurence
```

```python
# reverse
L = [2,1,5,7,6,3,4]
# permenently reverses
L.reverse()
print(L)
```

```python
# sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L)) # temporary sortin
print(L)
L.sort() # permanent sorting
print(L)
```

```python
# copy -> Shallow copy
L = [2,1,5,7,6,3,4]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```

List Comprehension

List Comprehension provides a concise way of creating list (short cut to create a list)

newlist = [expression for item in iterable if condition == True]

Advatages of List Comprehension
*   More time-efficient and space-efficient than loops.
*   Require fever lines of code
*   Transforms iterative statement into a formulas





```python
# Add 1 to 10 numbers to a list - common way
L =[]

for i in range(1,11):
  L.append(i)
print(L)
```

```python
# Comprehension way
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector - common way
v = [2,3,4]
s = -3

x = []
for i in v:
  x.append(i*s)
print(x)
```

```python
# Comprehension way
v = [2,3,4]
s = -3

[s*i for i in v]
```

```python
# Add squares - Comprehension way
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5 == 0]
```

```python
# find languages which starts with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('p')]
```

```python
# Nested if with Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'
[fruits for fruits in my_fruits if fruits in basket if fruits.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested list comprehension
[[i*j for i in range(1,4) ]for j in range(1,4)]
```

```python
# Cartesian products -> List Comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

2 ways to traverse the list


*   itemwise
*   indexwise




```python
# itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
# indexwise
L = [1,2,3,4]

for i in range(0,len(L)):
  print(i)
```

Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```
