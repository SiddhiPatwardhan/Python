### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

1-D set of numbers can be called series.

### Series from lists

The numbering mentioned is called the index of item

Series object consists of 2 things index and value

Datatype here is object

````python
# string
country = ['India','Pakistan','USA','Nepal','Srilanka']

pd.Series(country)
````

````python
# integers
runs = [13,24,56,78,100]

pd.Series(runs)
````

````python
runs_ser = pd.Series(runs)
````

````python
# custom index
marks = [67,57,89,100]

subjects = ['maths','english','science','hindi']

pd.Series(marks,index=subjects)
````

````python
# setting a name
marks = pd.Series(marks,index=subjects,name='Marks')
marks
````

### Series from dict

````python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name='Marks')
marks_series
````
